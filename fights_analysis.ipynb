```python
import wget
import zipfile

url = 'https://www.kaggle.com/api/v1/datasets/download/neelagiriaditya/ufc-datasets-1994-2025'
wget.download(url, 'data.zip')

with zipfile.ZipFile('./data.zip', 'r') as zip_ref:
  zip_ref.extractall('./data')
```

```python
import pandas as pd

csv_file = pd.read_csv('data/fight_details.csv')
csv_file.describe()
```

```python
csv_file.head()
```

```python
events = csv_file.groupby(csv_file['event_name'])[['r_name', 'r_total_str_landed', 'r_total_str_atmpted', 'b_name', 'b_total_str_landed', 'b_total_str_atmpted']]

# Fither with higher number of strikes attempted
last_event = events.get_group('UFC Fight Night: Imavov vs. Borralho')
red_max_strikes, blue_max_strikes = last_event[['r_total_str_atmpted', 'b_total_str_atmpted']].max()

result = ''
if red_max_strikes >= blue_max_strikes:
  result = last_event[last_event['r_total_str_atmpted'] == red_max_strikes][['r_name']]
else:
  result = last_event[last_event['b_total_str_atmpted'] == blue_max_strikes][['b_name']]

result
```
