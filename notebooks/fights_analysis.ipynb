```python
import wget
import zipfile

url = 'https://www.kaggle.com/api/v1/datasets/download/neelagiriaditya/ufc-datasets-1994-2025'
wget.download(url, 'data.zip')

with zipfile.ZipFile('../data.zip', 'r') as zip_ref:
  zip_ref.extractall('../data')
```

```python
import pandas as pd

csv_file = pd.read_csv('data/fight_details.csv')
csv_file.describe()
```

```python
csv_file.head()
```

```python
events = csv_file.groupby(csv_file['event_name'])[['r_name', 'r_total_str_landed', 'r_total_str_atmpted', 'r_sig_str_landed', 'r_sig_str_atmpted', 'b_name', 'b_total_str_landed', 'b_total_str_atmpted', 'b_sig_str_landed', 'b_sig_str_atmpted']]

# Fither with higher number of strikes attempted
last_event = events.get_group('UFC Fight Night: Imavov vs. Borralho')

# Strikes dataframe
names = pd.concat([last_event['b_name'], last_event['r_name']], ignore_index=True)
strikes_atmptd = pd.concat([last_event['b_total_str_atmpted'], last_event['r_total_str_atmpted']], ignore_index=True)
strikes_landed = pd.concat([last_event['b_total_str_landed'], last_event['r_total_str_landed']], ignore_index=True)
sig_strikes_atmptd = pd.concat([last_event['b_sig_str_atmpted'], last_event['r_sig_str_atmpted']], ignore_index=True)
sig_strikes_landed = pd.concat([last_event['b_sig_str_landed'], last_event['r_sig_str_landed']], ignore_index=True)
strikes_effectivity_result = (strikes_landed * 100)/strikes_atmptd
sig_strikes_effectivity_result = (sig_strikes_landed * 100)/sig_strikes_atmptd

df = pd.DataFrame({'Name': names, 
                   'Strikes Attempted': strikes_atmptd,
                   'Strikes Landed': strikes_landed,
                   'Strikes Effectivity': strikes_effectivity_result,
                   'Sign. Strikes Attempted': sig_strikes_atmptd,
                   'Sign. Strikes Landed': sig_strikes_landed,
                   'Sign. Strikes Effectivity': sig_strikes_effectivity_result
                  })

df
```

```python
# Fighter with most strikes attempted
df[df['Strikes Attempted'] == df['Strikes Attempted'].max()]
```

```python
# Scatter plot with strikes attempted per each fighter

import matplotlib.pyplot as plt

plt.scatter(df['Strikes Attempted'], df['Name'])
plt.show()
```

```python
# Horizontal bar plot with strikes landed per each fighter
colors = ["skyblue"] * len(df['Name'])

colors[df['Strikes Attempted'].idxmax()] = "orange"
plt.barh(df['Name'], df['Strikes Attempted'], color=colors)
plt.show()
```

```python
strikes_atmptd_mean = df['Effectivity'].mean()

print(strikes_atmptd_mean)

df[df['Effectivity'] < strikes_atmptd_mean-10]
```

```python
mean = df['Strikes Attempted'].mean()
std = df['Strikes Attempted'].std()

# Compute z-scores
z_scores = (df['Strikes Attempted'] - mean) / std

# Select outliers (|z| > 3)
df[z_scores.abs() > 3]
```
